```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import pandas as pd
import util
%matplotlib inline
```

```python
reader = csv.reader(open("../data/individual_trend_data/tensorflow_trend.csv", "rt"), delimiter=",") #read the .csv
next(reader, None) #skip the first row, that's just the column names
x = list(reader)
#print (x)
years=[]
count=[]
for i in x:
    years.append(i[0])
    count.append(i[1])
x_vals = np.array(years).astype(np.int)
y_vals = np.array(count).astype(np.float)
x_vals.sort()
y_vals.sort()
#print (x_vals)
#print (y_vals)
```

```python
plt.figure(figsize=(20,8))
plt.plot(x_vals, y_vals, color='orange')
plt.xlabel('Years')
plt.ylabel('Repo Count')
plt.title('TensorFlow usage over the years')
plt.show()
```

```python
data = pd.read_csv("files.csv")
X = data['f0_']
```

```python
X_dict = dict()
for i in range(0,len(X)):
    if X[i] in X_dict:
        X_dict[X[i]] += 1
    else:
        X_dict[X[i]] = 1

SortX = sorted(X_dict.items(), key=lambda s: s[0])
```

```python
x_vals = []
y_vals = []
for i in range(0,len(SortX)):
    x_vals.append(SortX[i][0])
    y_vals.append(SortX[i][1])
```

```python
y_pos = np.arange(len(x_vals))
plt.figure(figsize=(20,8))
plt.bar(y_pos, y_vals, align='center', alpha=0.5)
plt.xticks(y_pos, x_vals)
plt.ylabel('No of Repos')
plt.xlabel('Files Count')
plt.title('No of repos for a file count')
plt.ylim(0,500)
plt.xlim(0,100)
plt.show()
```
