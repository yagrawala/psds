```python
import copy
import util
import numpy as np
%matplotlib inline
```

```python
folder_path = "../data/language_data/"
langs = {}
for i in range(2008, 2019):
    vals = util.read_2D_values(folder_path + str(i) + ".csv")
    vals = vals[1:]
    for j in range(0, len(vals)):
        if vals[j][0] not in  langs.keys():
            langs[vals[j][0]] = {}
        langs[vals[j][0]][i] = vals[j][1]

# print( langs )
for lang in langs:
    temp = []
    for i in range(2008, 2019):
        if i in langs[lang].keys():
            temp.append( langs[lang][i] )
        else:
            temp.append( 0 )
    langs[lang] = temp
```

```python
lang_list = ["JavaScript", "Perl", "Java", "Python",
             "Ruby", "PHP", "C++", "CSS", "C#", "C"
            ]
#              "R", "SQL", "HTML", "Go", "Shell",
#              "Objective-C", "Scala", 
#              "Swift", "TypeScript"]
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan']
plots = []
counter = 1
processed = {}
for lang in langs:
    temp = np.array( langs[lang] ).astype(np.float)
    if lang in lang_list:
        processed[lang] = temp
        plots.append( util.getYPlotObj(temp, "line", label=lang, color=colors[counter%len(colors)]) )
        counter += 1

X_range = np.array( range(2008, 2019) )
util.plot_graph(X_range, plots, showLegend=True)
```

```python
total = np.zeros( len(processed["Java"]) )
for lang in lang_list:
    for i in range(0, len(processed["Java"])):
        total[i] += processed[lang][i]

creation_perc = copy.deepcopy( processed )
for lang in lang_list:
    for i in range(0, len(processed["Java"])):
        creation_perc[lang][i] = processed[lang][i] / total[i] * 100

plots_2 = []
counter = 1
for lang in lang_list:
    plots_2.append( util.getYPlotObj(creation_perc[lang], "line", label=lang, color=colors[counter%len(colors)]) )
    counter += 1
util.plot_graph(X_range, plots_2, showLegend=True)
```

```python
population = copy.deepcopy( processed )
for lang in lang_list:
    for i in range(1, len(population["Java"])):
        population[lang][i] += population[lang][i-1]

total_2 = np.zeros( len(population["Java"]) )
for lang in lang_list:
    for i in range(0, len(population["Java"])):
        total_2[i] += population[lang][i]

population_perc = copy.deepcopy( processed )
for lang in lang_list:
    for i in range(0, len(population["Java"])):
        population_perc[lang][i] = population[lang][i] / total_2[i] * 100

plots_3 = []
counter = 1
for lang in lang_list:
    plots_3.append( util.getYPlotObj(population_perc[lang], "line", label=lang, color=colors[counter%len(colors)]) )
    counter += 1
util.plot_graph(X_range, plots_3, showLegend=True)
```

```python
plots = []
counter = 1
for lang in lang_list:
    print( lang )
    [average_error, errors, predictions] = util.make_predictions(    
                                                            creation_perc[lang],
                                                            method="ewma",
                                                            ewma_factor=0.7,
                                                            season_factor=144,
                                                            ar_factor=2,
                                                            test_start_idx=1
                                                        )
    pre = creation_perc[lang][0]
    predictions = [pre] + predictions
    plots.append( util.getYPlotObj(creation_perc[lang], "line", label=lang, color=colors[counter%len(colors)]) )
    plots.append( util.getYPlotObj(predictions, "line", color=colors[counter%len(colors)], ls="-.") )
    counter += 1
util.plot_graph(X_range, plots, showLegend=True)
```

```python
def getMAPE(Y, Y_pred):
	MAPE_val = 0
	count = 0
	for i in range(0,len(Y)):
		if Y[0] != 0:
			map_val = abs((Y[0] - Y_pred[0])/Y[0])
			MAPE_val += map_val
			count += 1

	MAPE = (MAPE_val*100/count)
	return MAPE

plots = []
counter = 1
errors_foo = []
for lang in lang_list:
    print( lang )
    [average_error, errors, predictions] = util.make_predictions(    
                                                            population_perc[lang],
                                                            method="ewma",
                                                            ewma_factor=0.75,
                                                            season_factor=144,
                                                            ar_factor=2,
                                                            test_start_idx=1
                                                        )
    pre = population_perc[lang][0]
    predictions = [pre] + predictions
    mape = getMAPE(population_perc[lang], predictions )
    print( mape )
    errors_foo.append( mape )
    plots.append( util.getYPlotObj(population_perc[lang], "line", label=lang, color=colors[counter%len(colors)]) )
    print( population_perc[lang][-1])
    plots.append( util.getYPlotObj(predictions, "line", color=colors[counter%len(colors)], ls="-.") )
    counter += 1
# X_R = np.array( range(2008, 2021) )
util.plot_graph(X_range, plots, X_label="Year", Y_label="Percentage %", showLegend=True)
avg_error = util.getMean( errors_foo )
print( avg_error )
```
