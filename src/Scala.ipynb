```python
import numpy as np
import matplotlib.pyplot as plt
import csv
from scipy.stats import beta
```

```python
first_reader = csv.reader(open("C:/Users/Aditi/Desktop/Prob-Stats/Individual_language_data/Scala_trend.csv", "rt"), delimiter=",") #read the .csv
next(first_reader, None) #skip the first row, that's just the column names
x = list(first_reader)

years=[]
repo_count=[]

for i in x:
    years.append(i[0])
    repo_count.append(i[1])

x_vals = np.array(years).astype(np.int)
y_vals = np.array(repo_count).astype(np.float)

predicted=[0]*len(years)
for i in range(1,len(years)):
    predicted[i]=(repo_count[i-1])

residuals=[0]*len(years)
for i in range(1,len(years)):
    residuals[i]=int(repo_count[i])-int(predicted[i])
    
for i in range(len(repo_count)):
    repo_count[i]=int(repo_count[i])

for i in range(len(predicted)):
    predicted[i]=int(predicted[i])

x_vals=x_vals[:-3]
residuals=residuals[:-3]
predicted=predicted[:-3]
repo_count=repo_count[:-3]

print("residuals= ",residuals)
print("repo_count= ",repo_count)
print("predicted= ",predicted)
print("x_vals= ",x_vals)

```

```python
plt.figure(figsize=(20,8))

plt.plot(x_vals, residuals, color='blue', label='Residuals',linestyle='solid',linewidth=6)
plt.plot(x_vals, repo_count, color='red',label='Actual Values',linestyle='dotted',linewidth=6)
plt.plot(x_vals, predicted, color='green',label='Predicted Values',linestyle='dashed',linewidth=6)

plt.xticks(x_vals,fontsize=18)
plt.yticks(fontsize=18)
plt.xlabel('Years',fontsize=20)
plt.ylabel('Repo Count',fontsize=20)
plt.title('Scala Trend',fontsize=20)
plt.legend(loc='upper right')
plt.show()
 
```
